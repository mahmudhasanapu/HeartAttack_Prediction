#Data preprocessing
null removing by doing median.

```python
#education
median=df['education'].median()
df.loc[df['education'].isnull() == True, 'education'] = df['education'].median()

#cigsPerDay
median=df['cigsPerDay'].median()
df.loc[df['cigsPerDay'].isnull() == True, 'cigsPerDay'] = df['cigsPerDay'].median()

#BPMeds
median=df['BPMeds'].median()
df.loc[df['BPMeds'].isnull() == True, 'BPMeds'] = df['BPMeds'].median()

#totChol
median = df['totChol'].median()
df.loc[df['totChol'].isnull() == True, 'totChol'] = df['totChol'].median()

#BMI
median = df['BMI'].median()
df.loc[df['BMI'].isnull() == True, 'BMI'] = df['BMI'].median()

#heartRate
median = df['heartRate'].median()
df.loc[df['heartRate'].isnull() == True, 'heartRate'] = df['heartRate'].median()

#glucose
median = df['glucose'].median()
df.loc[df['glucose'].isnull() == True, 'glucose'] = df['glucose'].median()

print(df.isnull().sum())
```

#Data_preprocessing using minmaxscaler

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
output = ['TenYearCHD']
data1 = scaling.fit_transform(df.drop(output,axis=1))
data1
```

```python
data2 = pd.DataFrame(data1, columns=['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds','prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP','diaBP', 'BMI', 'heartRate', 'glucose'])
data2
```

```python
output1 = df[['TenYearCHD']]
output1
```

```python
new_data = pd.concat([data2, output1], axis=1)
new_data
```

```python
x=new_data.drop('TenYearCHD',axis=1)
y=new_data[['TenYearCHD']]
x,y
```

#Model building and training
model_LogisticRegression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=2)
x_train,x_test,y_train,y_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
d1=pd.DataFrame(y_pred,columns=['y_pred'])
d1
```

```python
model.score(x_test,y_test)
```

```python
output=y_test.copy()
output['y_pred']=y_pred
output
```

#Performence evulation of the Model
confusion_matrix,accuracy_score,f1_score,classification_report,precision_score,recall_score

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,classification_report,precision_score,recall_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
print(classification_report(y_test,y_pred))


```
